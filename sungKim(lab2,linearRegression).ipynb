```python
import tensorflow.compat.v1 as tf #downgrade tensorflow to 1.0
tf.disable_v2_behavior()
x_train = [1,2,3]
y_train = [1,2,3]

W = tf.Variable(tf.random.normal([1]), name = 'weight')
b = tf.Variable(tf.random.normal([1]), name = 'bias')
#Our hypothesis Xw+b
hypothesis = x_train * W + b # w= 1 b= 0이 되는 게 정답
```

```python
# cost/loss funtion
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
#Launch the graph in a session.
sess = tf.Session()
#Initializers global variables in the graph
sess.run(tf.global_variables_initializer())

#Fit the line
for step in range(2001):
    sess.run(train)
    if step%20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
#Use placeholders
X = tf.placeholder(tf.float32, shape = [None])
Y = tf.placeholder(tf.float32, shape = [None]) #1d array no limited elements
#Hypothesis
hypothesis =  X * W + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
#Launch the graph in a session.
sess = tf.Session()
#Initializers global variables in the graph
sess.run(tf.global_variables_initializer())

#Fit the line with new training data
for step in range(2001):
    cost_val, W_val, b_val, train_val = sess.run([cost, W, b, train], feed_dict={X:[1, 2, 3, 4, 5], Y:[2.1, 3.1, 4.1, 5.1, 6.1]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
#w = 1, b = 1.1
```

```python
#test
print(sess.run(hypothesis, feed_dict={X:[5]}))
print(sess.run(hypothesis, feed_dict={X:[2.5]}))
print(sess.run(hypothesis, feed_dict={X:[1.5, 3.5]}))
```
