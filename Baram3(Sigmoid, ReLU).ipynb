activation functions

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def relu_function(x):
    return np.maximum(0,x)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = relu_function(x)
plt.plot(x,y)
plt.ylim(-5, 5)
plt.show()
```
