```python
def AND(x1,x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2 
    if tmp <= theta:
        return 0
    else:
        return 1
```

```python
def AND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND(0,0)
```

```python
AND(0,1)
```

```python
AND(1,0)
```

```python
AND(1,1)
```

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0),NAND(0,1),NAND(1,0),NAND(1,1))
```

```python
 def XOR(x1,x2):
        s1 = NAND(x1,x2)
        s2 = OR(x1,x2)
        y = AND(s1,s2)
        return y
```

```python
print(XOR(0,0), XOR(0,1), XOR(1,0), XOR(1,1))
```

```python
def XXOR(x1,x2):
    s2 = NAND(x1,x2)
    s4 = NAND(x1,s2)
    s5 = NAND(x2,s2)
    y = NAND(s4,s5)
    return y
```

```python
print(XXOR(0,0), XXOR(0,1),XXOR(1,0),XXOR(1,1))
```
